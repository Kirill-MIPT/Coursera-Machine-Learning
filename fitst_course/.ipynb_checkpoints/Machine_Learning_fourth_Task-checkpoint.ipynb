```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

#сгенерируйте 1000 выборок объёма n и постройте гистограммы beta распределений их выборочных средних, n = 50
from scipy.stats import beta
rv_beta = sts.beta(2,2)
sample_beta = rv_beta.rvs(1000)
m = 1/2.    # теоретич. мат ожидание beta распределения m =  alfa/(alfa+betta)
d = np.sqrt(1/20.) # среднеквадратич отклонение d = sqrt((alfa*betta)(alfa+betta)**2*)
# генерим функцию норм распределения с необходимыми параметрами m,d и функцию для выборочных средних
def clt(sample,sub_sample_size,n_sub_samples=1000):
    means = []
    for i in range(n_sub_samples):
        sub_sample = sample[np.random.randint(0,len(sample),sub_sample_size)]
        means.append(np.mean(sub_sample))
        
    mu = np.mean(means)
    sigma = np.std(means,ddof=1)
    SE = d/np.sqrt(sub_sample_size) # стандартная ошибка среднего
    
    norm_rv = sts.norm(m,SE)
    
    print 'Теоретическое среднее vs приближенное: {:.2f} vs {:.2f}'.format(m,mu) #печать теоретич значений 
    print 'Теоретическое стандартное откл. vs приближенное: {:.2f} vs {:.2f}'.format(SE,d)
    print 'Приближенное среднее 95% доверительный интервал = {:.2f} +/- {:.2f}'.format(mu,2*SE)
    
    plt.hist(means,normed=True,label='actual') # тут строятся графики 
    plt.plot(x,norm_rv.pdf(x),label='theoretical')
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
clt(sample_beta,5) ## запускаем функцию выборка = 5 
```
